We can find the number of characters in a string by using the len() method.
Try this:

#3. Finally Booleans

*   Booleans are truth values that represent binary states like on or off
*   The two Boolean values are True and False
*   Capitalization is Important




```
year_is_2020 = True
print(year_is_2020)
```

The 'not' keyword negates a Boolean value.

```
#We are finished toasting
toaster_is_on = False
toaster_is_off = not toaster_is_on
print(toaster_is_on)
print(toaster_is_off)
```
